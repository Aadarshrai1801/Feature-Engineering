### Handling Imbalanced Dataset
###### 1. Up Sampling
###### 2. Down Sampling

```python
import numpy as np
import pandas as pd
```

```python
# Set the random seed for reproducibility
np.random.seed(123)
```

```python
# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# Create my dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    "feature 1" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "feature 2" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "target" : [0] * n_class_0
})

class_1 = pd.DataFrame({
    "feature 1" : np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "feature 2" : np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "target" : [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df.head()
```

```python
df["target"].value_counts()
```

### 1. Up Sampling

```python
df_minority = df[df["target"] == 1]
df_majority = df[df["target"] == 0]
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled = resample(df_minority, replace = True, n_samples = len(df_majority),random_state = 42)
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
df_upsampled["target"].value_counts()
```

### 2. Down Sampling

```python
# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

# Create my dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    "feature 1" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "feature 2" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "target" : [0] * n_class_0
})

class_1 = pd.DataFrame({
    "feature 1" : np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "feature 2" : np.random.normal(loc = 2, scale = 1, size = n_class_1),
    "target" : [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df["target"].value_counts()
```

```python
df_majority = df[df["target"] == 0]
df_minority = df[df["target"] == 1]
```

```python
df_majority_downsampled = resample(df_majority, replace = False, n_samples = len(df_minority),random_state = 42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled["target"].value_counts()
```
