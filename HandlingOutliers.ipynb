### Handling Outliers using Python

### 5 Number Summary and Box Plot

```python
import numpy as np
import seaborn as sns
```

```python
list_marks = [22,56,45,87,99,45,34,27,85,34,99,98,33]

minimum, Q1, median, Q3, maximum = np.quantile(list_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum
```

```python
maximum
```

```python
Q1
```

```python
Q3
```

```python
median
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR
```

```python
lower_fence
```

```python
higher_fence
```

```python
sns.boxplot(list_marks)
```
